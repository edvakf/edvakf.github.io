```python
import pandas as pd

athlete_events = pd.read_csv('./rawdata/athlete_events.csv')
athlete_events.replace(replacement, inplace=True)
athlete_events.replace({'URS': 'RUS'}, inplace=True)
athlete_events.replace({'FRG': 'DEU'}, inplace=True)

noc_regions = pd.read_csv('./rawdata/noc_regions.csv')

df = pd.merge(left=athlete_events, right=noc_regions, how='left', on='NOC')
df = df[df['Year'] >= 1960]
df = df[df['Season'] == 'Summer']
```

```python
df = df.dropna(subset=['Medal'])
df = df.filter(items=['Year', 'NOC', 'region', 'Sport', 'Event', 'Medal'])
df = df.drop_duplicates(subset=['NOC', 'Year', 'Sport', 'Event'])
```

```python
df.to_csv('./medals.csv', index=False)
```

```python
df2 = df.groupby(['Year', 'NOC']).agg(medalcount=('Medal', 'count'))
df2
```

```python
df2.to_csv('./medals-count-by-year.csv', index_label=['Year', 'NOC'])
```
